Given the root of a binary tree, return the inorder traversal of its nodes' values.

**Note:**

**Constraints:**

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# with stack
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        
        node_stack = []
        res = []
        while node_stack or root:
            if root:
                node_stack.append(root)
                root = root.left
            else:
                node = node_stack.pop()
                res.append(node.val)
                root = node.right
        return res
        
```

```python
# with dfs
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        
        res = []
        
        def dfs(node):
            if not node:
                return
            
            if node.left:
                dfs(node.left)
                
            res.append(node.val)
            
            if node.right:
                dfs(node.right)
                
        dfs(root)
        
        return res
```

Given the root of a binary tree, check whether it is a mirror of itself (i.e., symmetric around its center).

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# BFS with queue
class Solution:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        if not root:
            return True
        
        def isMirrorTree(p, q):

            from collections import deque
            queue = deque([(p, q)])
            
            while queue:
                nodeP, nodeQ = queue.popleft()
                
                if not nodeP and not nodeQ:
                    continue
                elif None in (nodeP, nodeQ):
                    return False
                else:
                    if nodeP.val != nodeQ.val:
                        return False
                    queue.append((nodeP.left, nodeQ.right)) # This is where the mirror checking happens.
                    queue.append((nodeP.right, nodeQ.left))
                    
            return True
        
        return isMirrorTree(root.left, root.right)
```

```python
# Lecture solution
class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if root == None:
            return True
        S = []

        #create a mirror tree of the original tree
        #so we push the root twice
        S.append(root)
        S.append(root)
        while (len(S) != 0):
            #for each node, we have two copies and we call one L and another R.
            L = S.pop()
            R = S.pop()
            #check the L and R
            if L.val != R.val:
                return False
            #check the children of L and R 
            if L.left != None:
                if R.right == None:
                    return False
                else:
                    #push L's left and R's right to the stack
                    S.append(L.left)
                    S.append(R.right)
            else: #L.left == None
                if R.right != None:
                    return False

            if L.right != None:
                if R.left == None:
                    return False
                else:
                    #push L's right and R's left to the stack
                    S.append(L.right)
                    S.append(R.left)
            else:
                if R.left != None:
                    return False
        return True
```

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def binaryTreePaths(self, root: Optional[TreeNode]) -> List[str]:
        rL = []
        if not root:
            return rL
        
        stack = [(root, '')]
        while stack:
            node, node_path = stack.pop()
            if not node.left and not node.right:
                rL.append(node_path + str(node.val))
            if node.right:
                stack.append((node.right, node_path + str(node.val) + "->"))
            if node.left:
                stack.append((node.left, node_path + str(node.val) + "->"))
        
        return rL
                
```

```python
# Lecture solution

class Solution(object):
        
    def binaryTreePaths(self, root):
        """
        :type root: TreeNode
        :rtype: List[str]
        """

        if not root:
            return []
        res = []
        #We need to remebmer the path for all the node, so when we push root to stack, we keep track of the path so far as well.
        stack = [(root, "")]
        while stack:
            node, path = stack.pop()
            #at the end of a path, a node has no children
            if not node.left and not node.right:
                res.append(path+str(node.val))
            if node.right:
                stack.append((node.right, path+str(node.val)+"->"))
            if node.left:
                stack.append((node.left, path+str(node.val)+"->"))
        return res
        
```

### Example: Find Largest Value in Each Tree Row

**Description:**

Given the root of a binary tree, return an array of the largest value in each row of the tree **(0-indexed)**.

**Note:**


**Example 1:**

![alt_text](images/leetcode515_ex1_largest_e1.jpeg)

Input: root = [1,3,2,5,3,null,9]

**Constraints:**

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def largestValues(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        
        hashmap = collections.defaultdict(int)
        
        def dfs(node, depth):
            if not node:
                return
            
            if depth not in hashmap:
                hashmap[depth] = node.val
            else:
                hashmap[depth] = max(hashmap[depth], node.val)
                
            if node.left:
                dfs(node.left, depth+1)
                
            if node.right:
                dfs(node.right, depth+1)
                
        dfs(root, 0)
        return hashmap.values()
```

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def largestValues(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        
        res = []
        
        queue = collections.deque([root])
        
        while queue:
            row_max = -2**31 - 1
            for _ in range(len(queue)):
                node = queue.popleft()
                row_max = max(row_max, node.val)
                
                if node.left:
                    queue.append(node.left)
                
                if node.right:
                    queue.append(node.right)
                    
            if row_max > -2**31 - 1:
                res.append(row_max)
        
        return res
```

```python
# Lecture solution
class Solution(object):
    def largestValues(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root: 
            return []

        #Using queue, which is BFS
        from collections import deque
        
        rL= []
        queue = deque([root])
 
        while queue:
            size = len(queue)
            level_max = float('-inf')
            #in each for loop, we have added all the nodes in one layer. 
            for i in range(size):
                check = queue.popleft()
                level_max = max(level_max, check.val)
                if check.left:
                    queue.append(check.left)
                if check.right:
                    queue.append(check.right)
                    
            rL.append(level_max)

        return rL
```

Given a n-ary tree, find its maximum depth.

The maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.

Nary-Tree input serialization is represented in their level order traversal, each group of children is separated by the null value (See examples).

**Note:**


**Example 1:**

<img src='images/leetcode559_ex1_narytreeexample.png' width=600 height=600 />

Input: root = [1,null,3,2,4,null,5,6]

Output: 3


**Example 2:**

<img src='images/leetcode559_ex2_sample_4_964.png' width=600 height=600 />

Input: root = [1,null,2,3,4,5,null,null,6,7,null,8,null,9,10,null,null,11,null,12,null,13,null,null,14]

Output: 5


**Constraints:**

- The total number of nodes is in the range [0, $10^{4}$].
- The depth of the n-ary tree is less than or equal to 1000.

```python
# Definition for a Node.
class Node:
    def __init__(self, val=None, children=None):
        self.val = val


class Solution:
    def maxDepth(self, root: 'Node') -> int:
        depth = 0
        if not root:
            return depth
        
        from collections import deque
        queue = deque([root])
        
        while queue:
            for _ in range(len(queue)):
                node = queue.popleft()
                if node.children:
                    for child in node.children:
                        queue.append(child)
            depth += 1
        
        return depth
    
                
                
```

### Exercise:  N-ary Tree Preorder Traversal
[589. N-ary Tree Preorder Traversal](https://leetcode.com/problems/n-ary-tree-preorder-traversal/)

**Description:**

Given the root of an n-ary tree, return the preorder traversal of its nodes' values.

Nary-Tree input serialization is represented in their level order traversal. Each group of children is separated by the null value (See examples)

**Note:**


**Example 1:**

Output: [1,3,5,6,2,4]


**Example 2:**

Input: root = [1,null,2,3,4,5,null,null,6,7,null,8,null,9,10,null,null,11,null,12,null,13,null,null,14]

Output: [1,2,3,6,7,11,14,4,8,12,5,9,13,10]

**Constraints:**

- The number of nodes in the tree is in the range [0, $10^{4}$].
- 0 <= Node.val <= 104
- The height of the n-ary tree is less than or equal to 1000.

```python
# Definition for a Node.
class Node:
    def __init__(self, val=None, children=None):
        self.val = val
        self.children = children
# DFS with stack
# BFS is not good for this as the question requires preorder traversal 
class Solution:
    def preorder(self, root: 'Node') -> List[int]:
        rL = []
        if not root:
            return rL
        
        stack = [root]
        while stack:
            node = stack.pop()
            rL.append(node.val)
            if node.children:
                for child in node.children[::-1]:
                    stack.append(child)
        return rL
```

```python
# Lecture solution

class Solution(object):
    def preorder(self, root):
        """
        :type root: Node
        :rtype: List[int]
        """
        if not root:
            return []
        
        res = []
        stack = [root]
        while stack:
            node = stack.pop()
            res.append(node.val)
            #note, we need to visit the children reversely. 
            for i in range(len(node.children)-1, -1, -1):
                stack.append(node.children[i])
        return res
```
