**Note:**


**Example 1:**

<img src="https://github.com/jason13123/python_coding_basics/blob/master/images/leetcode94_ex1_inorder_1.jpeg?raw=1" width=100 height=100>

**Example 3:**

**Constraints:**

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# with stack
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        
        node_stack = []
        res = []
        while node_stack or root:
            if root:
                node_stack.append(root)
                root = root.left
            else:
                node = node_stack.pop()
                res.append(node.val)
                root = node.right
        return res
        
```

### Exercise:  Title
[100. Same Tree](https://leetcode.com/problems/same-tree/)

**Description:**

Given the roots of two binary trees p and q, write a function to check if they are the same or not.

Two binary trees are considered the same if they are structurally identical, and the nodes have the same value.

**Note:**


**Example 1:**

![alt_text](images/same_tree_ex1.jpeg)

Input: p = [1,2,3], q = [1,2,3]

Output: true

**Example 2:**

![alt_text](images/same_tree_ex2.jpeg)

Input: p = [1,2], q = [1,null,2]

Output: false

**Example 3:**

![alt_text](images/same_tree_ex3.jpeg)

Input: p = [1,2,1], q = [1,1,2]

Output: false

**Constraints:**

- The number of nodes in both trees is in the range [0, 100].
- -$10^{4}$ <= Node.val <= $10^{4}$

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# BFS
class Solution:
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
        if not p and not q:
            return True
        elif not p or not q:
            return False
        
        from collections import deque
        
        queue_p = deque([p])
        queue_q = deque([q])
        
        while queue_p or queue_p:
            if len(queue_p) != len(queue_p):
                return False
            
            size = len(queue_p)
            for _ in range(size):
                node_p = queue_p.pop()
                node_q = queue_q.pop()
                
                if node_p.val != node_q.val:
                    return False
                if bool(node_p.left) != bool(node_q.left):
                    return False
                if bool(node_p.right) != bool(node_q.right):
                    return False
                
                if node_p.left:
                    queue_p.append(node_p.left)
                if node_p.right:
                    queue_p.append(node_p.right)
                
                if node_q.left:
                    queue_q.append(node_q.left)
                if node_q.right:
                    queue_q.append(node_q.right)
                    
        return True
                
```

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

        
# https://www.youtube.com/watch?v=LOM8Gd7m-kc&t=2s

'''
Idea: depth first search manner. Recursively traverse both subtree, but again: 
- Traverse the right subtree first
- Then traverse left subtree

In this recursive manner:
 - base case: node is null --> return
 - Do we need to return anything back on top? No actually, at each level this time, you will probably need to somehow still save all the element on the same level, and then basically get the first element of that level entually. A Hashmap may come into rescue here.
    + The key will be the level you wanna save
    + The value is the list of element on that level, with the first one being the rightmost one
- another key thing here is at every recursive function, you'll need to need to remember the state you are at. You state is defined by (node, level)
'''

class Solution:
    def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
        from collections import defaultdict
        res = defaultdict(int)
        
        def dfs(node, depth):
            if not node:
                return
            if depth not in res:
                res[depth] = node.val
            
            if node.right:
                dfs(node.right, depth+1)
            
            if node.left:
                dfs(node.left, depth+1)
            

        dfs(root, 0)
        
        return res.values()
```

The encoding rule is: k[encoded_string], where the encoded_string inside the square brackets is being repeated exactly k times. Note that k is guaranteed to be a positive integer.

You may assume that the input string is always valid; there are no extra white spaces, square brackets are well-formed, etc. Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, k. For example, there will not be input like 3a or 2[4].

The test cases are generated so that the length of the output will never exceed $10^{5}$.


**Note:**

A leaf is a node with no children.

**Example 1:**

Input: s = "3[a]2[bc]"

Output: "aaabcbc"

**Example 2:**

Input: s = "3[a2[c]]"

**Example 3:**

Input: s = "2[abc]3[cd]ef"

Output: "abcabccdcdcdef"

**Constraints:**

- 1 <= s.length <= 30
- s consists of lowercase English letters, digits, and square brackets '[]'.
- s is guaranteed to be a valid input.
- All the integers in s are in the range [1, 300].

```python
class Solution:
    def decodeString(self, s: str) -> str:
        num = 0
        current = ''
        stack = []
        
        for c in s:
            if c.isdigit():
                num = num * 10 + int(c)
            elif c == '[':
                stack.append([num, current])
                num = 0
                current = ''
            elif c == ']':
                tempNum, tempCurrent = stack.pop()
                current = tempCurrent + tempNum * current
            else:
                current += c
        
        return current
```

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def largestValues(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        
        hashmap = collections.defaultdict(int)
        
        def dfs(node, depth):
            if not node:
                return
            
            if depth not in hashmap:
                hashmap[depth] = node.val
            else:
                hashmap[depth] = max(hashmap[depth], node.val)
                
            if node.left:
                dfs(node.left, depth+1)
                
            if node.right:
                dfs(node.right, depth+1)
                
        dfs(root, 0)
        return hashmap.values()
```

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def largestValues(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        
        res = []
        
        queue = collections.deque([root])
        
        while queue:
            row_max = -2**31 - 1
            for _ in range(len(queue)):
                node = queue.popleft()
                row_max = max(row_max, node.val)
                
                if node.left:
                    queue.append(node.left)
                
                if node.right:
                    queue.append(node.right)
                    
            if row_max > -2**31 - 1:
                res.append(row_max)
        
        return res
```

```python
# Lecture solution

class Solution(object):
    def maxDFS(self, rTupleList, tempD, root):
        if not root.left and not root.right:
            rTupleList[tempD][0] = max(root.val, rTupleList[tempD][0])
            return
        if root.left:
            if rTupleList[-1][1] < tempD+1:
                rTupleList.append([root.left.val, tempD+1])
            rTupleList[tempD+1][0] = max(root.left.val, rTupleList[tempD+1][0])
            
            self.maxDFS(rTupleList, tempD+1, root.left)
            
        if root.right:
            if rTupleList[-1][1] < tempD+1:
                rTupleList.append([root.right.val, tempD+1])
            rTupleList[tempD+1][0] = max(root.right.val, rTupleList[tempD+1][0])
            
            self.maxDFS(rTupleList, tempD+1, root.right)

    def largestValues(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root:
            return []

        rTupleList = [[root.val, 0]]
        self.maxDFS(rTupleList, 0, root)
        
        return list(map(lambda x:x[0], rTupleList))
```
